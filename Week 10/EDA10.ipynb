```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload the cleaned dataset file
uploaded = files.upload()

# Load the dataset
df = pd.read_excel('Cleaned_DATASET.xlsx')

# Display dataset information
print("Dataset Info:")
print(df.info())

# Display first few rows
print("\nFirst Few Rows:")
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Count the values in the Persistency_Flag column
persistency_counts = df['Persistency_Flag'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
persistency_counts.plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    labels=['Non-Persistent', 'Persistent'],
    colors=['skyblue', 'orange']
)
plt.title('Distribution of Persistency_Flag')
plt.ylabel('')  # Remove y-axis label for better aesthetics
plt.show()
```

```python
# Bar chart for Gender distribution
gender_counts = df['Gender'].value_counts()
gender_counts.plot(kind='bar', color=['lightblue', 'orange'])
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

# Persistency vs Gender
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender', hue='Persistency_Flag', palette='pastel')
plt.title("Persistency by Gender", fontsize=16)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title='Persistency Flag')
plt.show()

# Persistency vs Race
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Race', hue='Persistency_Flag', palette='pastel')
plt.title("Persistency by Race", fontsize=16)
plt.xlabel("Race", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Persistency Flag')
plt.show()

# Persistency vs Region
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Region', hue='Persistency_Flag', palette='pastel')
plt.title("Persistency by Region", fontsize=16)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title='Persistency Flag')
plt.show()

# Persistency vs Age_Bucket
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Age_Bucket', hue='Persistency_Flag', palette='pastel')
plt.title("Persistency by Age Bucket", fontsize=16)
plt.xlabel("Age Bucket", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title='Persistency Flag')
plt.xticks(rotation=45)
plt.show()
```
